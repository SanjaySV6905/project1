```python
data=original_data[["Severity","Start_Time","State","Visibility(mi)","Weather_Condition"]].copy()
```

```python
data.columns = ["Accident Severity", "Accident DateTime", "State", "Visibility", "Weather Condition"]
```

```python
data=data.dropna(subset=["Visibility","Weather Condition"])
```

```python
data["Visibility"]=data["Visibility"].round(0).astype(int)
```

```python
data["Weather Condition"]=data["Weather Condition"].str.strip().str.title()
```

```python
def categorize_weather(i):
    i = str(i).lower()
    
    if 'clear' in i or 'fair' in i:
        return 'Clear'
    elif 'cloud' in i or 'overcast' in i:
        return 'Cloudy'
    elif 'rain' in i or 'shower' in i or 'drizzle' in i:
        return 'Rain'
    elif 'snow' in i or 'sleet' in i or 'blowing snow' in i:
        return 'Snow'
    elif 'fog' in i or 'haze' in i or 'mist' in i:
        return 'Fog'
    elif 'thunder' in i or 'storm' in i or 't-storm' in i:
        return 'Thunderstorm'
    else:
        return 'Other'

data['Weather Condition'] = data['Weather Condition'].apply(categorize_weather)
```

```python
data['Accident DateTime'] = pd.to_datetime(data['Accident DateTime'], format='mixed')
```

```python
data['Day'] = data['Accident DateTime'].dt.day
data['Month'] = data['Accident DateTime'].dt.month
```
